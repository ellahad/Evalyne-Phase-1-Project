<h2>Extract Transform Load</h2>

```python
# convert date objects into pandas datetime
budget['release_date'] = pd.to_datetime(budget['release_date'])
budget.head()
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies.head()
```

```python
# Drop null values
gross.dropna(inplace = True)
gross.info()
```

```python
# Create new dataframe from names
name = names[['nconst' , 'primary_name' , 'known_for_titles']]
name.dropna(inplace=True)
name.info()
```

```python
principals.drop(['job' , 'characters'], axis = 1 , inplace = True)
principals.info()
```

```python
#Dropping null values in title dataframe
title.dropna(subset='primary_title',inplace=True)
title.info()
```

```python
#convert strings to lowercase
movies.original_title = movies.original_title.str.lower()
title.primary_title = title.primary_title.str.lower() 
budget.movie = budget.movie.str.lower()
```

```python
# Find values that match in movies and title dataframe
matching_values_movies_titles = movies[movies['original_title'].isin(title['primary_title'])]
matching_values_titles_movies = title[title['primary_title'].isin(movies['original_title'])]
movie = matching_values_movies_titles
titles = matching_values_titles_movies 
# Find values that don't match in movies and title dataframe
non_matching_values_movies = movies[~movies['original_title'].isin(title['primary_title'])]
non_matching_values_title = title[~title['primary_title'].isin(movies['original_title'])]

print("Matching values:")
print(movie)

print("\nValues in df1 that don't match df2:")
print(non_matching_values_movies)

print("\nValues in df2 that don't match df1:")
print(non_matching_values_title)
```

```python
# Drop title column
movie.drop(['title'], axis = 1 , inplace = True)
titles.drop(['original_title'], axis = 1 , inplace = True)
#changing original_title and primary_title to title
movie.rename(columns={'original_title': 'title'}, inplace=True)
titles.rename(columns={'primary_title': 'title'}, inplace=True)
#combining the movies and title dataframes
#movie_titles = movie.merge(titles, how = 'outer', on = movie.title)
#movie_titles
titles
#movie
```

```python
movie
```

```python
#combining the movies and title dataframes
movie_titles = movie.merge(titles, how = 'outer', on = 'title')
movie_titles
```

```python
# Find values that match between movies and budget
matching_values_movies_budget = movie_titles[movie_titles.title.isin(budget.movie)]
print("Matching values:")
print(matching_values_movies_budget)
```

```python
#changing movie to title
budget.rename(columns={'movie': 'title'}, inplace=True)
budget
```

```python
films = matching_values_movies_budget.merge(budget, how = 'inner', on = 'title')
films['domestic_gross']=films['domestic_gross'].str.replace('$', '')
films['domestic_gross']=films['domestic_gross'].str.replace(',', '')
films['domestic_gross']=films['domestic_gross'].astype('int')
films['worldwide_gross']=films['worldwide_gross'].str.replace('$', '')
films['worldwide_gross']=films['worldwide_gross'].str.replace(',', '')
films['worldwide_gross']=films['worldwide_gross'].astype('int')
films['production_budget']=films['production_budget'].str.replace('$', '')
films['production_budget']=films['production_budget'].str.replace(',', '')
films['production_budget']=films['production_budget'].astype('int')
films.info()
```

<h2>Exploratory Data Analysis</h2>

```python
roles = principals.category.unique()
roles
```

```python
actor = principals[principals['category'] == 'actor']
actor = actor.merge(films, how = 'inner', on = 'tconst')
actors=actor.groupby('nconst').sum({'domestic_gross','production_budget','worldwide_gross'}).reset_index()
actors = actors.merge(name, how = 'left', on = 'nconst')
actors.sort_values('worldwide_gross', ascending=False, inplace=True)
top = 30
actors.head()
#Dropping null values in title dataframe
actors.dropna(subset='primary_name',inplace=True)
production=np.array(actors['production_budget'])[:top]
production =tuple(production)
domestic=np.array(actors['domestic_gross'])[:top]
domestic=tuple(domestic)
worldwide = np.array(actors['worldwide_gross'])[:top]
worldwide =tuple(worldwide)
#Plot bar graph
fig, ax = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

ax[0].bar(tuple(np.array(actors['primary_name'])[:top]), production, color='r')
ax[0].set_ylabel('Production budget $')
ax[1].bar(tuple(np.array(actors['primary_name'])[:top]), domestic)
ax[1].set_ylabel('Domestic gross $')
ax[2].bar(tuple(np.array(actors['primary_name'])[:top]), worldwide, color='g')
ax[2].set_ylabel('Worldwide gross $')
plt.xticks(rotation = 90)
ax[0].set_title('Actors budget and gross')

# Save plot
plt.savefig("actors budget and gross.png")
```
